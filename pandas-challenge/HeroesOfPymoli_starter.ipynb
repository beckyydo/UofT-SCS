```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Calculate Total Number of Players
total_players = len(purchase_data["SN"].unique())

#Display in summmary form
pd.DataFrame([{"Total Players": total_players }])
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Individual Calculations
#Number of Unique Items
num_items = len(purchase_data["Item Name"].unique())

#Average Price
average_price = round(purchase_data["Price"].mean(),2)

#Number of Purchases
num_purchases = len(purchase_data)

#Total Revenue
ttl_revenue = purchase_data["Price"].sum()

#Combine Results into Dataframe
purchasing_analysis=pd.DataFrame([{"Number of Unique Items": num_items,
                     "Average Price": average_price,
                     "Number of Purchases": num_purchases,
                     "Total Revenue": ttl_revenue}])

#Format Average Price and Total Revenue
purchasing_analysis['Average Price'] =  purchasing_analysis['Average Price'].map("${:.2f}".format)
purchasing_analysis['Total Revenue'] =  purchasing_analysis['Total Revenue'].map("${:,.2f}".format)

#Display Dataframe
purchasing_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Take only SN and Gender From Dataframe
gender = pd.DataFrame(purchase_data, columns = ["SN","Gender"])
gender = gender.drop_duplicates()

#Count of Players' Gender
gender = gender.groupby("Gender").count()

#Percentage of Players' Gender
gender["Percentage of Players"] = gender["SN"]/total_players*100

#Rename Columns
gender = gender.rename(columns = {"SN": 'Total Count'})

#Format Percentage of Players
gender['Percentage of Players'] = gender['Percentage of Players'].map("{:.2f}%".format)

#Display summary
gender
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Group by Gender
purchase_gender = purchase_data.groupby("Gender")

#Individual calculations
#Total Count
purchase_analysis = pd.DataFrame({'Purchase Count': purchase_gender['Purchase ID'].count()})

#Average Purchase
purchase_analysis['Average Purchase'] = round(purchase_gender["Price"].mean(),2)

#Total Purchase
purchase_analysis['Total Purchase Value'] = purchase_gender["Price"].sum()

#Average Total Purchase per Person
purchase_analysis['Avg Total Purchase per Person'] = round(purchase_analysis['Total Purchase Value']/gender['Total Count'],2)

#Format
purchase_analysis['Average Purchase'] =purchase_analysis['Average Purchase'].map("${:.2f}".format)
purchase_analysis['Total Purchase Value'] =purchase_analysis['Total Purchase Value'].map("${:.2f}".format)
purchase_analysis['Avg Total Purchase per Person'] =purchase_analysis['Avg Total Purchase per Person'].map("${:.2f}".format)

#Display Summary
purchase_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Determine min and max in order to determine bins
purchase_data['Age'].min()
purchase_data['Age'].max()

#Set bin range and label
bins = [0, 9, 14, 19, 24, 29, 34, 39,48]
bin_labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

purchase_data['Bin'] = pd.cut(purchase_data['Age'], bins, labels = bin_labels)
purchase_age = purchase_data[['SN','Bin']].drop_duplicates()

#Groupby Dataframe by Bin
purchase_age = purchase_age.groupby('Bin')

#Calculate Count of players under Bin
purchase_age = purchase_age.count()

#Calculate percentage of players
purchase_age['Percentage of Players'] = purchase_age['SN']/total_players*100

#Format percentage of players
purchase_age['Percentage of Players'] = purchase_age['Percentage of Players'].map("{:.2f}%".format)
purchase_age = purchase_age.rename(columns={'SN':'Total Count'})

#Display Summary
purchase_age
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Group data using same bins from 'Age Demographic'
age_groupby = purchase_data.groupby('Bin')

#Calculation of columns
purchasing_analysis = age_groupby['Purchase ID'].count()
average_price = age_groupby['Price'].mean()
ttl_purchase = age_groupby['Price'].sum()

#Merge Groupby Dataframes
purchase_analysis = pd.merge(purchasing_analysis, average_price, on = 'Bin')
purchase_analysis = pd.merge(purchase_analysis, ttl_purchase, on = 'Bin')

#Add column Avg Total Purchase per Person
purchase_analysis['Avg Total Purchase per Person'] = purchase_analysis['Price_y']/purchase_analysis['Purchase ID']

#Rename Header
purchase_analysis = purchase_analysis.rename(columns={'Purchase ID': 'Purchase Count', 
                                                      'Price_x': 'Average Purchase Price',
                                                     'Price_y': 'Total Purchase Value'})

#Format Columns
purchase_analysis['Average Purchase Price'] = purchase_analysis['Average Purchase Price'].map("${:.2f}".format)
purchase_analysis['Total Purchase Value'] = purchase_analysis['Total Purchase Value'].map("${:.2f}".format)
purchase_analysis['Avg Total Purchase per Person'] = purchase_analysis['Avg Total Purchase per Person'].map("${:.2f}".format)

#Display Summary
purchase_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Group by user
user_groupby = purchase_data.groupby("SN")

#Calculate Columns
user_count = user_groupby['Purchase ID'].count()
user_avg_price = user_groupby['Price'].mean()
user_ttl_price = user_groupby['Price'].sum()

#Merge Dataframe
top_spender = pd.merge(user_count, user_avg_price, on ='SN')
top_spender = top_spender.merge(user_ttl_price, on = 'SN')

#Sort Data before Formatting
top_spender = top_spender.sort_values(by='Price_y', ascending = False)

#Rename Header
top_spender = top_spender.rename(columns={'Purchase ID': 'Purchase Count', 
                                          'Price_x': 'Average Purchase Price',
                                          'Price_y': 'Total Purchase Value'})

#Format Columns
top_spender['Average Purchase Price'] = top_spender['Average Purchase Price'].map("${:,.2f}".format)
top_spender['Total Purchase Value'] = top_spender['Total Purchase Value'].map("${:,.2f}".format)

#Display preview of summary
top_spender.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Select only columns needed and group by ID and Item Name
item_df = purchase_data[['Item ID','Item Name','Price']]
popular_groupby = item_df.groupby(['Item ID','Item Name'])

#Perform individual calculation
popular_count = popular_groupby['Price'].count()
popular_price = popular_groupby['Price'].mean()
popular_ttl = popular_groupby['Price'].sum()

#Merge Dataframes
popular = pd.merge(popular_count,popular_price, on =['Item ID','Item Name'])
popular = popular.merge(popular_ttl, on = ['Item ID','Item Name'])

#Sort data before formatting
most_popular = popular.sort_values('Price_x', ascending = False)

#Rename Columns
most_popular = most_popular.rename(columns = {'Price_x':'Purchase Count',
                                   'Price_y':'Item Price',
                                   'Price':'Total Purchase Value'})

#Format Item Price and Total Purchase Value
most_popular['Item Price'] = most_popular['Item Price'].map("${:.2f}".format)
most_popular['Total Purchase Value'] = most_popular['Total Purchase Value'].map("${:.2f}".format)

#Display preview of summary
most_popular.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#Sort data in descending order
most_profitable = popular.sort_values('Price', ascending = False)

#Rename Columns
most_profitable = most_profitable.rename(columns = {'Price_x':'Purchase Count',
                                   'Price_y':'Item Price',
                                   'Price':'Total Purchase Value'})

#Format Item Price and Total Purchase Value
most_profitable['Item Price'] = most_profitable['Item Price'].map("${:.2f}".format)
most_profitable['Total Purchase Value'] = most_profitable['Total Purchase Value'].map("${:.2f}".format)

#Display preview of summary
most_profitable.head()
```
